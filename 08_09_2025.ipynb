```python
#hybird inheritance
class A :
  def Hi(self):
    print("Good Morning")

class B(A):
  def Hello(self):
    print("Good Afternoon")

class C(A):
  def Bye(self):
    print("Good Evening")

class D(B,C):
  def Good (self):
    print("Good Night")

obj = D()
obj.Hi()
obj.Hello()
obj.Bye()
obj.Good()

```

```python
#Run time polymoprhism

class Animal:
  def sound(self):
    print("Animal makes a sound")

class Dog(Animal):
  def sound(self):
    print("Bow bow bow")

class Cat(Animal):
  def sound(self):
    print("Meow meow........")

class Cow(Animal):
  def sound(self):
    print("Ambooooooooooo")

d = Dog()
d.sound()

a = Animal()
a.sound()

c = Cat()
c.sound()

d = Cow()
d.sound()
```

```python
# Method Overloding(Compile time polymorphism)

class Match :
  def add(self,a,b=0,c=0):
    return a*b-c

obj = Match()
print(obj.add(7))
print(obj.add(6,4))
print(obj.add(4,7,3))
```

```python

#Operator Overloading

class Book:
  def __init__(self,pages):
    self.pages = pages
  def _add_(self, other):
    return self.pages + other.pages

b1 = Book(100)
b2 = Book(200)
print("Total Pages, b1 + b2")
```

```python
class Dictionary:
    def __init__(self):
     #Store words in a dictionary (key = word,value = meaning)
      self.words = {}

    def add_word(self,word,meaning):
     # prevent duplicate keys
      if word in self.words:
         print(f"'{words}' already exists in dictionary. Duplicate not allowed!")

      else:
          self.words[word] = meaning
          print(f"'{word}' added successfully!")

    def search_word(self,word):
     #search for a word and display result in table format
      if word in self.words:
         print("/n=====Search Resulrt =====")
         print(f"{'Word':<15} | {'Meaning'}")
         print("-" * 40)
         print(f"{word:<15} | {self.words[word]}")
         print("-" * 40)
      else:
         print(f"'{word}' not found in dictionary.")
    def show_all(self):
    # Show all words alphabetically
     if not self.words:
        print("Dictionary is empty.")
     else:
        print("\n===== My Dictionary (Alphabetical Order) =====")
        print(f"{'Word':<15} | {'Meaning'}")
        print("-" * 40)
        for word in sorted(self.words.keys()):  #Sort keys alphabetically
            print(f"{word:<15} | {self.words[word]}")
        print("-" * 40)

    def update_word(self, word,new_meaning):
    # Update word if it exists
     if word in self.words:
        self.words[word] = new_meaning
        print(f"'{word}' updated successfully! ")
     else:
        print(f"'{word}' not found. Cannot update.")


    def main():
      my_dict = Dictionary()

      while True:
         print("\n===== Digital Dictionary Menu =====")
         print("1. Add Word")
         print("2. Search Word")
         print("3. Show All Words")
         print("4. Update Word Meaning")
         print("5. Exit")

         choice = input("Enter choice: ")

         if choice == "1":
            word = input("Enter word: ")
            meaning = input("Enter meaning: ")
            my_dict.add_word(word,meaning)

         elif choice == "2":
            word = input("Enter word to search: ")
            my_dict.search_word(word)

         elif choice == "3":
            my_dict.show_all()

         elif choice == "4":
            word = input("Enter word to update: ")
            meaning = input("Enter new meaning: ")
            my_dict.update_word(word,meaning)

         elif choice == "5":
            print("Exiting Dictionary... Goodbye!")
            break
         else:
            print("Invalid choice.Try again.")

  #Run the progaram
  if __name__ == " __main__":
      main()


```
